```python
from sklearn.impute import SimpleImputer
df[num_cols] = SimpleImputer(strategy="median").fit_transform(df[num_cols])
df[cat_cols] = SimpleImputer(strategy="most_frequent").fit_transform(df[cat_cols])
```

## 2. Encoding Techniques

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cat_cols = df.select_dtypes(include="object").columns
cat_cols
```

```python
# One-Hot Encoding for small categories (< 5 unique values)
df_encoded = pd.get_dummies(df, columns=[col for col in cat_cols if df[col].nunique() <= 5],drop_first=True)
```

```python
df_encoded 
```

```python
# Label Encoding for large categories (> 5 unique values)
le = LabelEncoder()
for col in cat_cols:
    if col in df_encoded.columns and df_encoded[col].dtype == "object":
        df_encoded[col] = le.fit_transform(df_encoded[col])
```

## 3.Feature Engineering

```python
# New feature 1: Net capital gain
df_encoded["capital_diff"] = df_encoded["capital_gain"] - df_encoded["capital_loss"]
```

```python
df_encoded["capital_diff"]
```

```python
# New feature 2: Grouping working hours
df_encoded["hours_group"] = pd.cut(df_encoded["hours_per_week"],
                                   bins=[0, 20, 40, 60, 100],
                                   labels=["low", "medium", "high", "very high"])
```

```python
df_encoded["hours_group"]
```

```python
# Log transformation to reduce skewness
df_encoded["capital_gain_log"] = np.log1p(df_encoded["capital_gain"])
```

```python
df_encoded["capital_gain_log"]
```

## 4.Feature Selection

#### (a) Remove Outliers with Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(contamination=0.02, random_state=42)
```

```python
suchi = iso.fit_predict(df_encoded[num_cols])  # only numeric features
mask = suchi != -1
df_no_outliers = df_encoded[mask]
```

```python
print("Original shape:", df_encoded.shape)
print("After removing outliers:", df_no_outliers.shape)
```
